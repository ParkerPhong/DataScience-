````python
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]
print(np.polyfit(X,Y,1))
````

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
   
    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    m = numer / denum
    b = ybar - m * xbar

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(b,m))

    return m, b
    

# solution
m, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
# ANSWER (~3 lines)
plt.scatter(X,Y)
yhat = [b + m * xi for xi in X]
plt.plot(X, yhat)

plt.show()
````

# 11. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

```
# Expected Output:
(30.476190476190474, 19.539561081583383, 41.41281987079756)
```

````python


def mean_confidence_interval(data, confidence=0.95):
    
    # Calculate m, h
    # ANSWER (~3 lines)
    
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1 )
    return m, m-h, m+h

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))

````
